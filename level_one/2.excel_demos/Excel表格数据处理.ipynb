## 行索引

```python
import pandas as pd

data_loc = r'./resources/销售数据.xlsx'
data = pd.read_excel(data_loc)

# 读取第一行的值
# 索引第一行的值，行标签是“0”
data_0 = data.loc[0]
data_0
```

```python
data
```

```python
print(data_0)
```

```python
# 读取第二列全部值
data_1 = data.loc[ : ,"大类编码"]
data_1
```

```python
# 读取第1行，第B列对应的值
data_2 = data.loc[ 1, "大类编码"]
data_2
```

```python
data_3 = data.loc[ 1:3, "大类编码":"中类名称"]
data_3
```

```python
data_4 = data.loc[ data["销售数量"] > 10] 
data_4
```

```python
# 切片操作
data_5 = data.loc[ data["销售数量"] > 10, ["小类编码","小类名称"]]
data_5
```
